# text cleaning of captions:
- converting all sentences to lower case
- removing all numbers and punctuation
- removing words of length 1

```python
def text_cleaning(sent):
    sent = sent.lower()   #converting to lower case
    sent = re.sub("[^a-zA-Z]+"," ",sent) # removing numbers and punctuation
    sent = sent.split()
    
    sent = [w for w in sent if len(w)>1] # removing words of length 1 
    sent = " ".join(sent)
    return sent
```

## Performing text cleaning of captions

```python
for key in descriptions.keys():
    for i in range(len(descriptions[key])):
        descriptions[key][i] = text_cleaning(descriptions[key][i])
```

## Filtering words from vocabulary based on a threshold frequency

```python
# creating a dictionary of {word:num_of_occurences}
import collections
counter = collections.Counter(total_words)
freq_cnt = dict(counter)
```

## taking threshold 8

```python
sorted_freq_cnt = sorted(freq_cnt.items(),reverse=True,key = lambda x:x[1]) # sorting freq_cnt in descending order
threshold = 8
vocab_words = [word[0] for word in sorted_freq_cnt if word[1] > threshold ]
```

## creating word_to_idx and idx_to_word dictionaries

```python
word_to_idx = {}
idx_to_word = {}

for ix,word in enumerate(vocab_words):
    word_to_idx[word] = ix + 1
    idx_to_word[ix+1] = word 
```

```python
# Loading Resnet50 model with pretrained imagenet weights
# This model will be used to encode images in (2048,) vectors, which will be used to represent an image for input.
res_model = ResNet50(weights = 'imagenet',input_shape = (224,224,3))
res_model.summary()
```

```python
IMG_PATH = '/Users/mananmehta/Desktop/flickr8k/Flickr_Data/Images/'
```

```python
def preprocess_img(img):
    img = image.load_img(img,target_size = (224,224))
    img = image.img_to_array(img)
    img = np.expand_dims(img,axis = 0)
    #Normalisation
    img = preprocess_input(img)
    return img
```

```python
def encode_img(img):
    img = preprocess_img(img)
    feature_vector = new_res_model.predict(img)
    feature_vector = feature_vector.reshape((2048,))
    return feature_vector
```

# creating a custom data generator

### Each Datapoint will contain image vector and corresponding partial caption
<img src="images/Unknown-6.jpeg" width="800" /> ![]("images/Unknown-6.jpeg")

```python
def data_loader(train_descriptions,encoded_images,max_len ,batchsize,word_to_idx):
    X1,X2,Y = [],[],[]
    n = 0
    while True:
        for key in train_descriptions.keys():
            n = n+1
            photo = encoded_images[key+'.jpg']
            
            for cap in train_descriptions[key]:
                seq = [word_to_idx[w] for w in cap.split() if w in word_to_idx]
                for i in range(1,len(seq)):
                    xi = seq[0:i]
                    yi = seq[i]

                    xi = pad_sequences([xi],maxlen=max_len,value = 0,padding = 'post')[0]
                    yi = to_categorical([yi],num_classes=vocab_size)[0]

                    X1.append(photo)
                    X2.append(xi)
                    Y.append(yi)

                    
            if n== batchsize:
                yield [[np.array(X1),np.array(X2)],np.array(Y)]
                X1,X2,Y = [],[],[]
                n = 0
```

# Model Architecture


```python
input_img_features = Input(shape=(2048,))
img_feature1 = Dropout(0.3)(input_img_features)
img_feature2 = Dense(256,activation='relu')(img_feature1)
```

```python
input_cap_features = Input(shape = (maxlen,))
cap_features1 = Embedding(input_dim=vocab_size,output_dim=50,mask_zero=True)(input_cap_features)
cap_features2 = Dropout(0.3)(cap_features1)
cap_features3 = LSTM(256)(cap_features2)
```

```python
decoder_1 = add([img_feature2,cap_features3])
decoder_2 = Dense(256,activation='relu')(decoder_1)
output = Dense(vocab_size,activation='softmax')(decoder_2)

model = Model(inputs = [input_img_features,input_cap_features],outputs = output)
```

```python
model.summary()
```

```python
model.layers[2].set_weights([embedding_matrix]) # Using embedding_matrix as weights of embedding layer
model.layers[2].trainable = False  # We dont want to train embedding layer 
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam')
```

```python
epochs = 10
batch_size = 3
steps = len(train_descriptions)//batch_size
```

```python
for i in range(epochs):
    generator = data_loader(train_descriptions,encoded_images,maxlen,batch_size,word_to_idx)
    model.fit_generator(generator,epochs=1,steps_per_epoch=steps,verbose = 1)
    model.save_weights('./model_weights/model_'+ str(i) + '.h5')
```

```python
def predict_caption(photo):
    in_text = "startseq"
    
    for i in range(maxlen):
        sequence = [word_to_idx[w] for w in in_text.split() if w in word_to_idx]
        sequence = pad_sequences([sequence], maxlen=maxlen, padding='post')

        ypred =  model.predict([photo,sequence])
        ypred = ypred.argmax()
        word = idx_to_word[ypred]
        in_text+= ' ' +word
        
        if word =='endseq':
            break
        
        
    final_caption =  in_text.split()
    final_caption = final_caption[1:-1]
    final_caption = ' '.join(final_caption)
    
    return final_caption
```
